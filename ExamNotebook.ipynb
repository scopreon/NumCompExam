```python
fig, axs = plt.subplots(1,2)
fig.suptitle('Vertically stacked subplots')
fig.set_figwidth(10)
fig.set_figheight(5)
for ax in axs.flat:
    ax.set(xlabel='x-label', ylabel='y-label')
axs[1].set(xlabel='testst', ylabel='y-label')
axs[0].plot(a,b)
axs[1].plot(a,b)
```
